##F1 coding

```python
# Define color palette
COLOR_PALETTES = {
    'bar_gradient': plt.cm.YlGnBu,  # Yellow-Green-Blue gradient
    'mean_line': '#D95B43',  # Orange-red for mean line
    'threshold_line': '#00A676',  # Teal for threshold line
    'heatmap': 'coolwarm'  # Coolwarm color map for heatmap
}
```

```python
# Mean F1-score
MEAN_F1_SCORES = {
    'Bali': 0.9164,
    'Jawa': 0.9319,
    'Sunda': 0.9844,
    'Bali-Jawa': 0.9342,
    'Bali-Sunda': 0.9552,
    'Jawa-Sunda': 0.9643,
    'Bali-Jawa-Sunda': 0.9572
}
```

```python
def calculate_similarity_threshold(mean_f1, percentage=0.90):
    return percentage * mean_f1

def find_similar_characters(class_names, f1_scores, mean_f1, percentage=0.90):

    threshold = calculate_similarity_threshold(mean_f1, percentage)

    similar_chars = []
    for name, score in zip(class_names, f1_scores):
        if score < threshold:
            similar_chars.append({
                'character': name,
                'f1_score': score,
                'difference_from_threshold': threshold - score
            })

    similar_chars.sort(key=lambda x: x['difference_from_threshold'])

    return {
        'mean_f1_score': mean_f1,
        'threshold': threshold,
        'similar_characters': similar_chars
    }
```

```python
def visualize_f1_scores(class_names, f1_scores, title, mean_f1, percentage=0.90):

    # Calculate threshold
    threshold = calculate_similarity_threshold(mean_f1, percentage)

    # Prepare data for DataFrame
    df = pd.DataFrame({
        'Character': class_names,
        'F1-Score': f1_scores
    })

    # Create plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 16),
                                   gridspec_kw={'width_ratios': [3, 1]})  # Lext 3x wider than right

    # Sort data from highest to lowest F1-score
    df_sorted = df.sort_values('F1-Score', ascending=False)

    # Create color gradient based on F1-score
    norm = mcolors.Normalize(vmin=df_sorted['F1-Score'].min(), vmax=df_sorted['F1-Score'].max())
    colors = COLOR_PALETTES['bar_gradient'](norm(df_sorted['F1-Score']))

    # Bar plot with thinner bars to reduce crowding
    bars = ax1.bar(df_sorted['Character'], df_sorted['F1-Score'], color=colors, width=0.7)
    ax1.set_title(f'F1-Score for {title}', fontsize=23, fontweight='bold')
    ax1.set_xlabel('Character', fontsize=18)
    ax1.set_ylabel('F1-Score', fontsize=18)

    # Tick label font sizes
    ax1.tick_params(axis='y', labelsize=18)
    ax1.tick_params(axis='x', labelsize=18)

    # Legend font size
    ax1.legend(fontsize=18)



    # Add extra space at the bottom for labels
    fig.subplots_adjust(bottom=0.3)

    # Improved label rotation and spacing with smaller font
    plt.setp(ax1.get_xticklabels(),
             rotation=80,
             ha='right',
             rotation_mode='anchor',
             fontsize=12)

    # Ensure tick marks are visible and evenly spaced
    ax1.set_xticks(range(len(df_sorted)))
    ax1.set_xticklabels(df_sorted['Character'])

    # grid linws
    ax1.grid(axis='y', linestyle='--', alpha=0.7)

    # mean and threshold lines
    ax1.axhline(y=mean_f1, color=COLOR_PALETTES['mean_line'], linestyle='--',
                label=f'Mean F1-Score ({mean_f1:.4f})', linewidth=2)
    ax1.axhline(y=threshold, color=COLOR_PALETTES['threshold_line'], linestyle='--',
                label=f'Threshold ({threshold:.4f})', linewidth=2)
    ax1.legend(fontsize=12)

    # Boxplot for F1-score distribution
    box_df = pd.DataFrame({'F1-Score': f1_scores})
    sns.boxplot(data=box_df, x='F1-Score', ax=ax2)
    ax2.set_title(f'F1-Score Distribution of {title}', fontsize=16, fontweight='bold')
    ax2.set_xlabel('F1-Score', fontsize=17)

    # mean and threshold lines to boxplot
    ax2.axvline(x=mean_f1, color=COLOR_PALETTES['mean_line'], linestyle='--',
                label=f'Mean F1-Score ({mean_f1:.4f})', linewidth=2)
    ax2.axvline(x=threshold, color=COLOR_PALETTES['threshold_line'], linestyle='--',
                label=f'Threshold ({threshold:.4f})', linewidth=2)
    ax2.legend(fontsize=12)

    #overall spacing
    plt.tight_layout(rect=[0, 0.05, 1, 0.95])
    plt.show()
```

```python
def visualize_f1_scores_horizontal(class_names, f1_scores, title, mean_f1, percentage=0.90):
    # threshold
    threshold = calculate_similarity_threshold(mean_f1, percentage)

    # DataFrame
    df = pd.DataFrame({
        'Character': class_names,
        'F1-Score': f1_scores
    })

    # figure size
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, max(8, len(class_names) * 0.25)),
                                  gridspec_kw={'width_ratios': [2, 1]})  # Make left plot 3x wider than right

    # Sort data from highest to lowest F1-score
    df_sorted = df.sort_values('F1-Score', ascending=True)  # Ascending for horizontal bars

    # Create color gradient based on F1-score
    norm = mcolors.Normalize(vmin=df_sorted['F1-Score'].min(), vmax=df_sorted['F1-Score'].max())
    colors = COLOR_PALETTES['bar_gradient'](norm(df_sorted['F1-Score']))

    # Horizontal bar plot
    bars = ax1.barh(df_sorted['Character'], df_sorted['F1-Score'], color=colors, height=0.6)
    ax1.set_title(f'F1-Score for {title}', fontsize=16, fontweight='bold')
    ax1.set_ylabel('Character', fontsize=14)
    ax1.set_xlabel('F1-Score', fontsize=14)

    plt.setp(ax1.get_yticklabels(), fontsize=8)  # Reduced font size for better spacing

    # mean and threshold lines (vertical for horizontal bars)
    ax1.axvline(x=mean_f1, color=COLOR_PALETTES['mean_line'], linestyle='--',
                label=f'Mean F1-Score ({mean_f1:.4f})', linewidth=2)
    ax1.axvline(x=threshold, color=COLOR_PALETTES['threshold_line'], linestyle='--',
                label=f'Threshold ({threshold:.4f})', linewidth=2)
    ax1.legend(fontsize=12)

    # Boxplot for F1-score distribution
    box_df = pd.DataFrame({'F1-Score': f1_scores})
    sns.boxplot(data=box_df, x='F1-Score', ax=ax2)
    ax2.set_title(f'F1-Score Distribution of {title}', fontsize=16, fontweight='bold')
    ax2.set_xlabel('F1-Score', fontsize=14)

    # mean and threshold lines to boxplot
    ax2.axvline(x=mean_f1, color=COLOR_PALETTES['mean_line'], linestyle='--',
                label=f'Mean F1-Score ({mean_f1:.4f})', linewidth=2)
    ax2.axvline(x=threshold, color=COLOR_PALETTES['threshold_line'], linestyle='--',
                label=f'Threshold ({threshold:.4f})', linewidth=2)
    ax2.legend(fontsize=12)

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
def heatmap_similarity_comparison():

    # List of classifications and their scores
    classifications = [
        ("Bali", f1_scores_bali),
        ("Jawa", f1_scores_jawa),
        ("Sunda", f1_scores_sunda),
        ("Bali-Jawa", f1_scores_bali_jawa),
        ("Bali-Sunda", f1_scores_bali_sunda),
        ("Jawa-Sunda", f1_scores_jawa_sunda),
        ("Bali-Jawa-Sunda", f1_scores_bali_jawa_sunda)
    ]

    # Prepare data for heatmap
    similarity_matrix = np.zeros((len(classifications), len(classifications)))

    # Calculate similarity between classifications using min length
    for i, (name1, scores1) in enumerate(classifications):
        for j, (name2, scores2) in enumerate(classifications):
            # Use minimum length to ensure arrays have the same dimension
            min_length = min(len(scores1), len(scores2))
            scores1_trimmed = scores1[:min_length]
            scores2_trimmed = scores2[:min_length]

            # Use Pearson correlation to measure similarity
            similarity_matrix[i, j] = np.corrcoef(scores1_trimmed, scores2_trimmed)[0, 1]

    # Plot heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(similarity_matrix,
                annot=True,
                cmap=COLOR_PALETTES['heatmap'],
                center=0,
                cbar_kws={'label': 'Pearson Correlation'},
                xticklabels=[name for name, _ in classifications],
                yticklabels=[name for name, _ in classifications])
    plt.title('Heatmap Similaritas Antar Klasifikasi', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
def visualize_all_classifications(percentage=0.90):

    # List of classifications
    classifications = [
        ("Bali", class_names_bali, f1_scores_bali, MEAN_F1_SCORES['Bali']),
        ("Jawa", class_names_jawa, f1_scores_jawa, MEAN_F1_SCORES['Jawa']),
        ("Sunda", class_names_sunda, f1_scores_sunda, MEAN_F1_SCORES['Sunda']),
        ("Bali-Jawa", class_names_bali_jawa, f1_scores_bali_jawa, MEAN_F1_SCORES['Bali-Jawa']),
        ("Bali-Sunda", class_names_bali_sunda, f1_scores_bali_sunda, MEAN_F1_SCORES['Bali-Sunda']),
        ("Jawa-Sunda", class_names_jawa_sunda, f1_scores_jawa_sunda, MEAN_F1_SCORES['Jawa-Sunda']),
        ("Bali-Jawa-Sunda", class_names_bali_jawa_sunda, f1_scores_bali_jawa_sunda, MEAN_F1_SCORES['Bali-Jawa-Sunda'])
    ]

    # Visualize each classification
    for classification_name, names, scores, mean_f1 in classifications:
        print(f"\nVisualization of {classification_name}")

        # Threshold Calculation
        threshold = calculate_similarity_threshold(mean_f1, percentage)
        print(f"Threshold Calculation:")
        print(f"- Mean F1-Score: {mean_f1:.4f}")
        print(f"- Percentage: {percentage:.2f}")
        print(f"- Threshold = {percentage} × {mean_f1:.4f} = {threshold:.4f}")

        result = find_similar_characters(names, scores, mean_f1, percentage)

        print(f"\nSimilarity Classification Result:")
        print(f"Mean F1-Score: {result['mean_f1_score']:.4f} Threshold: {result['threshold']:.4f}")

        if result['similar_characters']:
            print("Characters with High Similarity:")
            for char in result['similar_characters']:
                print(f"- {char['character']}: F1-score {char['f1_score']:.4f}")

        # vertical bars for consistency
        visualize_f1_scores(names, scores, classification_name, mean_f1, percentage)
```

```python
# Main execution
if __name__ == "__main__":
    # Percentage threshold
    percentage = 0.90

    try:
        # Visualize F1-scores
        visualize_all_classifications(percentage)

        # heatmap
        heatmap_similarity_comparison()

    except Exception as e:
        print(f"Error occurred: {e}")
        # Print full traceback for debugging
        import traceback
        print(traceback.format_exc())
```

## Mean

```python
# DataFrame
df = pd.DataFrame(data)

# 1. Analisis Statistik Dasar
similarity_values = df['value'].values
mean = np.mean(similarity_values)
std_dev = np.std(similarity_values)

# threshold for outlier
mild_outlier_threshold = mean + 1.5 * std_dev
significant_outlier_threshold = mean + 2 * std_dev
extreme_outlier_threshold = mean + 3 * std_dev

print(f"Rata-rata similarity: {mean:.2f}%")
print(f"Standar deviasi: {std_dev:.2f}%")
print(f"Threshold outlier ringan (mean + 1.5*SD): {mild_outlier_threshold:.2f}%")
print(f"Threshold outlier signifikan (mean + 2*SD): {significant_outlier_threshold:.2f}%")
print(f"Threshold outlier ekstrem (mean + 3*SD): {extreme_outlier_threshold:.2f}%")
```

## Outlier

```python
# 2. Mengidentifikasi outlier
df['outlier_category'] = 'Normal'
df.loc[df['value'] >= mild_outlier_threshold, 'outlier_category'] = 'Mild'
df.loc[df['value'] >= significant_outlier_threshold, 'outlier_category'] = 'Significant'
df.loc[df['value'] >= extreme_outlier_threshold, 'outlier_category'] = 'Extreme'

# outlier berdasarkan kategori
outliers = df[df['outlier_category'] != 'Normal'].sort_values(by='value', ascending=False)
print("\n=== OUTLIER BERDASARKAN KATEGORI ===")
for category in ['Extreme', 'Significant', 'Mild']:
    category_outliers = outliers[outliers['outlier_category'] == category]
    if not category_outliers.empty:
        print(f"\n{category} Outliers (>{significant_outlier_threshold if category == 'Significant' else mild_outlier_threshold if category == 'Mild' else extreme_outlier_threshold}%):")
        for _, row in category_outliers.iterrows():
            print(f"{row['pair']}: {row['value']}% ({row['dataset']})")
```

```python
# 3. Analisis Frekuensi pair Karakter
def extract_base_pair(pair_string):
    if '(' in pair_string:
        return pair_string.split('(')[0].strip()
    return pair_string
```

```python
df['base_pair'] = df['pair'].apply(extract_base_pair)

# Calculate frekuensi outlier untuk setiap pair
outliers_df = df[df['outlier_category'] != 'Normal']
pair_frequency = {}

for _, row in outliers_df.iterrows():
    base_pair = row['base_pair']
    if base_pair in pair_frequency:
        pair_frequency[base_pair]['count'] += 1
        pair_frequency[base_pair]['datasets'].append(row['dataset'])
        pair_frequency[base_pair]['values'].append(row['value'])
        if row['value'] > pair_frequency[base_pair]['max_value']:
            pair_frequency[base_pair]['max_value'] = row['value']
    else:
        pair_frequency[base_pair] = {
            'count': 1,
            'datasets': [row['dataset']],
            'values': [row['value']],
            'max_value': row['value'],
            'script': row['script']
        }

# Mengonversi ke DataFrame
frequency_df = pd.DataFrame([
    {
        'pair': pair,
        'count': data['count'],
        'max_value': data['max_value'],
        'avg_value': sum(data['values']) / len(data['values']),
        'datasets': ', '.join(data['datasets']),
        'script': data['script']
    }
    for pair, data in pair_frequency.items()
])

# Mengurutkan berdasarkan nilai maksimum
frequency_df = frequency_df.sort_values(by=['count', 'max_value'], ascending=[False, False])

print("\n=== PASANGAN KARAKTER YANG PALING SERING MUNCUL SEBAGAI OUTLIER ===")
for _, row in frequency_df.iterrows():
    print(f"{row['pair']}: Muncul dalam {row['count']} dataset, nilai maks: {row['max_value']:.1f}%, nilai rata-rata: {row['avg_value']:.1f}%")
    print(f"  Dataset: {row['datasets']}")
```

```python
# Scatter plot dengan anotasi
plt.figure(figsize=(20, 12))  # Diperbesar untuk mengakomodasi anotasi
scripts = df['script'].unique()
colors = {
    'Bali': '#1E90FF',
    'Jawa': '#4CAF50',
    'Sunda': '#FF5722'
}

# Plot setiap script dengan anotasi
for script in scripts:
    script_data = df[df['script'] == script]
    scatter = plt.scatter(script_data.index, script_data['value'],
                         label=script,
                         color=colors[script],
                         marker='o',
                         alpha=0.7,
                         s=120)  # Sedikit diperbesar untuk visibilitas

    # Menambahkan anotasi untuk setiap titik
    for idx, row in script_data.iterrows():
        # Menggunakan offset yang bervariasi untuk menghindari overlap
        offset_x = 0.5
        offset_y = 0.3

        # Untuk titik dengan nilai tinggi, letakkan anotasi di atas
        if row['value'] > significant_outlier_threshold:
            offset_y = 1.0
            fontweight = 'bold'
            fontsize = 8
            bbox_props = dict(boxstyle="round,pad=0.3", facecolor='yellow', alpha=0.7)
        else:
            fontweight = 'normal'
            fontsize = 7
            bbox_props = dict(boxstyle="round,pad=0.2", facecolor='white', alpha=0.6)

        plt.annotate(row['pair'],
                    (idx, row['value']),
                    xytext=(offset_x, offset_y),
                    textcoords='offset points',
                    fontsize=fontsize,
                    fontweight=fontweight,
                    ha='left',
                    bbox=bbox_props,
                    arrowprops=dict(arrowstyle='->',
                                  connectionstyle='arc3,rad=0.1',
                                  alpha=0.6,
                                  color=colors[script]))

# Menambahkan threshold lines
plt.axhline(y=mild_outlier_threshold, color='#FFA07A', linestyle='--',
            label=f'Mild Outlier Threshold ({mild_outlier_threshold:.2f}%)', linewidth=2)
plt.axhline(y=significant_outlier_threshold, color='#FF6347', linestyle='--',
            label=f'Significant Outlier Threshold ({significant_outlier_threshold:.2f}%)', linewidth=2)
plt.axhline(y=extreme_outlier_threshold, color='#DC143C', linestyle='--',
            label=f'Extreme Outlier Threshold ({extreme_outlier_threshold:.2f}%)', linewidth=2)

plt.title('Scatter Plot of Characters-Likeness Rate (CLR) by Script Type with Character Pair Labels',
          fontsize=16, fontweight='bold')
plt.ylabel('Characters-Likeness Rate (CLR)', fontsize=12)
plt.xlabel('Index', fontsize=12)
plt.legend(fontsize=14, loc='upper left')
plt.grid(True, alpha=0.3)

# Menambahkan margin untuk anotasi
plt.subplots_adjust(right=0.85, top=0.9)
plt.tight_layout()
plt.show()


# Alternatif: Scatter plot dengan anotasi hanya untuk outlier (lebih bersih)
plt.figure(figsize=(18, 10))

for script in scripts:
    script_data = df[df['script'] == script]
    plt.scatter(script_data.index, script_data['value'],
               label=script,
               color=colors[script],
               marker='o',
               alpha=0.7,
               s=100)

# Anotasi hanya untuk outlier (significant dan extreme)
outlier_data = df[df['value'] >= mild_outlier_threshold]
for idx, row in outlier_data.iterrows():
    # Warna anotasi berdasarkan tingkat outlier
    if row['value'] >= extreme_outlier_threshold:
        bbox_color = '#FFB6C1'  # Light pink untuk extreme
        fontweight = 'bold'
        fontsize = 9
    elif row['value'] >= significant_outlier_threshold:
        bbox_color = '#FFFFE0'  # Light yellow untuk significant
        fontweight = 'bold'
        fontsize = 8
    else:
        bbox_color = '#E6E6FA'  # Light lavender untuk mild
        fontweight = 'normal'
        fontsize = 7

    plt.annotate(f"{row['pair']}\n({row['value']:.1f}%)",
                (idx, row['value']),
                xytext=(10, 10),
                textcoords='offset points',
                fontsize=fontsize,
                fontweight=fontweight,
                ha='left',
                bbox=dict(boxstyle="round,pad=0.3",
                         facecolor=bbox_color,
                         alpha=0.8,
                         edgecolor=colors[row['script']]),
                arrowprops=dict(arrowstyle='->',
                              connectionstyle='arc3,rad=0.2',
                              alpha=0.7,
                              color=colors[row['script']]))

plt.axhline(y=mild_outlier_threshold, color='#FFA07A', linestyle='--',
            label=f'Mild Outlier Threshold ({mild_outlier_threshold:.2f}%)', linewidth=2)
plt.axhline(y=significant_outlier_threshold, color='#FF6347', linestyle='--',
            label=f'Significant Outlier Threshold ({significant_outlier_threshold:.2f}%)', linewidth=2)
plt.axhline(y=extreme_outlier_threshold, color='#DC143C', linestyle='--',
            label=f'Extreme Outlier Threshold ({extreme_outlier_threshold:.2f}%)', linewidth=2)

plt.title('Scatter Plot with Outlier Annotations Only', fontsize=16, fontweight='bold')
plt.ylabel('Characters-Likeness Rate (CLR)', fontsize=12)
plt.xlabel('Index', fontsize=12)
plt.legend(fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


# Versi ketiga: Scatter plot interaktif dengan informasi lengkap
fig, ax = plt.subplots(figsize=(16, 10))

for script in scripts:
    script_data = df[df['script'] == script]
    scatter = ax.scatter(script_data.index, script_data['value'],
                        label=script,
                        color=colors[script],
                        marker='o',
                        alpha=0.7,
                        s=100)

# Menambahkan anotasi untuk titik tertinggi dari setiap script
for script in scripts:
    script_data = df[df['script'] == script]
    max_value_row = script_data.loc[script_data['value'].idxmax()]

    ax.annotate(f"Highest {script}:\n{max_value_row['pair']}\n({max_value_row['value']:.1f}%)",
                (max_value_row.name, max_value_row['value']),
                xytext=(15, 15),
                textcoords='offset points',
                fontsize=9,
                fontweight='bold',
                ha='left',
                bbox=dict(boxstyle="round,pad=0.4",
                         facecolor=colors[script],
                         alpha=0.3,
                         edgecolor=colors[script]),
                arrowprops=dict(arrowstyle='->',
                              connectionstyle='arc3,rad=0.3',
                              alpha=0.8,
                              color=colors[script],
                              lw=2))

ax.axhline(y=mild_outlier_threshold, color='#FFA07A', linestyle='--',
           label=f'Mild Outlier Threshold ({mild_outlier_threshold:.2f}%)', linewidth=2)
ax.axhline(y=significant_outlier_threshold, color='#FF6347', linestyle='--',
           label=f'Significant Outlier Threshold ({significant_outlier_threshold:.2f}%)', linewidth=2)
ax.axhline(y=extreme_outlier_threshold, color='#DC143C', linestyle='--',
           label=f'Extreme Outlier Threshold ({extreme_outlier_threshold:.2f}%)', linewidth=2)

ax.set_title('Scatter Plot with Highest Value Annotations per Script', fontsize=16, fontweight='bold')
ax.set_ylabel('Characters-Likeness Rate (CLR)', fontsize=12)
ax.set_xlabel('Index', fontsize=12)
ax.legend(fontsize=14)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
